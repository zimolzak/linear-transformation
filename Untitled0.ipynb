```
import matplotlib
from numpy import matrix, linalg, pi, sin, cos
import matplotlib.pyplot as plt
from random import uniform
```

```
unit_circle = []
theta = 0
while theta <= 2 * pi:
    unit_circle.append([sin(theta), cos(theta)])
    theta += 2 * pi / 32
u = matrix(unit_circle)
```

```
m_string = (str(uniform(-2,2)) + ' ' +
            str(uniform(-2,2)) + '; ' + 
            str(uniform(-2,2)) + ' ' +
            str(uniform(-2,2)))

#m_string = '1.2 0.9; 0.1 2'

a = matrix(m_string)
print(a)
evev = linalg.eig(a)
eigenvecs = evev[1].T # default is return vecs in columns, so I transpose it.

p = u * a.T

#### Plots

fig, ax = plt.subplots()

for i in range(len(eigenvecs)):
    xs = [0, eigenvecs[i,0]]
    ys = [0, eigenvecs[i,1]]
    plt.plot(xs, ys, 'r+-')
    
for i in range(len(u)):
    xs = [u[i,0], p[i,0]]
    ys = [u[i,1], p[i,1]]
    plt.plot(xs, ys, 'go-')

ax.set_aspect(1.0)
plt.show()
```
